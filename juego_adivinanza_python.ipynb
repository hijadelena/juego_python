```python
# Configuración de la conexión a Google Sheets
scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]

# Lista para almacenar resultados de partidas
resultados_partidas = []

# Función para solicitar una opción válida del menú
def obtener_opcion_menu():
    while True:
        try:
            opcion = int(input("Elige una opción: "))
            if 1 <= opcion <= 4:
                return opcion
            else:
                print("Opción no válida. Inténtalo de nuevo.")
        except ValueError:
            print("Por favor, ingresa un número.")

# Función para solicitar una opción de dificultad válida
def obtener_opcion_dificultad():
    while True:
        try:
            dificultad = int(input("Elige la dificultad:\n1. Fácil (20 intentos)\n2. Medio (12 intentos)\n3. Difícil (5 intentos)\n"))
            if 1 <= dificultad <= 3:
                return dificultad
            else:
                print("Opción no válida. Inténtalo de nuevo.")
        except ValueError:
            print("Por favor, ingresa un número.")

# Función para guardar resultados en la lista y posiblemente en Google Sheets
def guardar_resultado(nombre, resultado):
    resultados_partidas.append({"Nombre": nombre, "Resultado": resultado})
    print(f"Guardando en la lista: {nombre} - {resultado}")

# Función para mostrar estadísticas desde la lista
def mostrar_estadisticas(filtrar_por_usuario=None):
    print("\nEstadísticas de partidas:")

    if filtrar_por_usuario:
        estadisticas_filtradas = [partida for partida in resultados_partidas if partida["Nombre"] == filtrar_por_usuario]

        if not estadisticas_filtradas:
            print(f"No hay estadísticas para el usuario '{filtrar_por_usuario}'.")
            return

        for partida in estadisticas_filtradas:
            print(f"{partida['Nombre']}: {partida['Resultado']}")
    else:
        for partida in resultados_partidas:
            print(f"{partida['Nombre']}: {partida['Resultado']}")

# Función para generar un número aparentemente aleatorio basado en el tiempo actual
def generar_numero_aleatorio():
    import datetime
    import time
    seed = int(datetime.datetime.now().timestamp())
    return ((seed * 9301 + 49297) % 233280) / 233280.0

# Función para jugar en modo solitario
def jugar_modo_solitario():
    numero_a_adivinar = 0  # Número inicial
    intentos = 0

    print("Modo Solitario - Adivina el número entre 1 y 1000")
    dificultad = obtener_opcion_dificultad()

    if dificultad == 1:
        max_intentos = 20
        numero_a_adivinar = int(generar_numero_aleatorio() * 1000) + 1
    elif dificultad == 2:
        max_intentos = 12
        numero_a_adivinar = int(generar_numero_aleatorio() * 1000) + 1
    elif dificultad == 3:
        max_intentos = 5
        numero_a_adivinar = int(generar_numero_aleatorio() * 1000) + 1

    while intentos < max_intentos:
        suposicion = int(input("Tu suposición: "))
        intentos += 1

        if suposicion == numero_a_adivinar:
            print(f"¡Felicidades! Adivinaste el número en {intentos} intentos.")
            nombre_jugador = input("Ingresa tu nombre: ")
            guardar_resultado(nombre_jugador, "Ganó")
            return
        else:
            if suposicion < numero_a_adivinar:
                print("Mi número es mayor.")
            else:
                print("Mi número es menor.")

    print(f"Lo siento, te quedaste sin intentos. El número era {numero_a_adivinar}.")
    nombre_jugador = input("Ingresa tu nombre: ")
    guardar_resultado(nombre_jugador, "Perdió")

# Función para jugar en modo 2 jugadores
def jugar_modo_2_jugadores():
    numero_a_adivinar = int(input("Primer jugador, ingresa el número a adivinar (entre 1 y 1000): "))
    intentos = 0

    print("Modo 2 Jugadores - Segundo jugador, adivina el número.")
    dificultad = obtener_opcion_dificultad()

    if dificultad == 1:
        max_intentos = 20
    elif dificultad == 2:
        max_intentos = 12
    elif dificultad == 3:
        max_intentos = 5

    while intentos < max_intentos:
        suposicion = int(input("Tu suposición: "))
        intentos += 1

        if suposicion == numero_a_adivinar:
            print(f"¡Felicidades! Adivinaste el número en {intentos} intentos.")
            nombre_jugador = input("Ingresa tu nombre: ")
            guardar_resultado(nombre_jugador, "Ganó")
            return
        else:
            if suposicion < numero_a_adivinar:
                print("El número es mayor.")
            else:
                print("El número es menor.")

    print(f"Lo siento, te quedaste sin intentos. El número era {numero_a_adivinar}.")
    nombre_jugador = input("Ingresa tu nombre: ")
    guardar_resultado(nombre_jugador, "Perdió")

# Menú principal
while True:
    print("\nMenú Principal:")
    print("1. Partida modo solitario")
    print("2. Partida 2 Jugadores")
    print("3. Estadística")
    print("4. Salir")

    opcion = obtener_opcion_menu()

    if opcion == 1:
        jugar_modo_solitario()
    elif opcion == 2:
        jugar_modo_2_jugadores()
    elif opcion == 3:
        # Preguntar si el usuario quiere filtrar por nombre
        filtro_usuario = input("Ingresa el nombre de usuario para filtrar estadísticas (o deja en blanco para mostrar todas): ")
        if filtro_usuario:
            # Filtrar por nombre de usuario
            estadisticas_filtradas = [partida for partida in resultados_partidas if partida["Nombre"].lower() == filtro_usuario.lower()]
            if not estadisticas_filtradas:
                print(f"No hay estadísticas para el usuario '{filtro_usuario}'.")
            else:
                for partida in estadisticas_filtradas:
                    print(f"{partida['Nombre']}: {partida['Resultado']}")
        else:
            # Mostrar todas las estadísticas
            mostrar_estadisticas()
    elif opcion == 4:
        print("¡Hasta luego!")
        break





```

```python
import tkinter as tk
from tkinter import messagebox
from adivina_el_numero import jugar_modo_solitario, jugar_modo_2_jugadores, mostrar_estadisticas

class JuegoAdivinanzasUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Juego de Adivinanzas")

        # Botones para seleccionar modo de juego
        btn_solitario = tk.Button(root, text="Modo Solitario", command=self.jugar_solitario)
        btn_solitario.pack(pady=10)

        btn_2_jugadores = tk.Button(root, text="Modo 2 Jugadores", command=self.jugar_2_jugadores)
        btn_2_jugadores.pack(pady=10)

        btn_estadisticas = tk.Button(root, text="Estadísticas", command=self.mostrar_estadisticas)
        btn_estadisticas.pack(pady=10)

    def jugar_solitario(self):
        # Lógica para el modo solitario
        resultado = jugar_modo_solitario()
        self.mostrar_resultado(resultado)

    def jugar_2_jugadores(self):
        # Lógica para el modo 2 jugadores
        resultado = jugar_modo_2_jugadores()
        self.mostrar_resultado(resultado)

    def mostrar_estadisticas(self):
        # Lógica para mostrar estadísticas
        estadisticas = mostrar_estadisticas()
        self.mostrar_resultado(estadisticas)

    def mostrar_resultado(self, mensaje):
        messagebox.showinfo("Resultado", mensaje)

if __name__ == "__main__":
    root = tk.Tk()
    app = JuegoAdivinanzasUI(root)
    root.mainloop()
```
